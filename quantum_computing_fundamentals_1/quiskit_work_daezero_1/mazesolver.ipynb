```python
from qiskit import QuantumCircuit, transpile
from qiskit_aer import Aer

circuit = QuantumCircuit(3, 3)

circuit.h([0, 1, 2])

circuit.x(1)                
circuit.h(2)
circuit.ccx(0, 1, 2)      
circuit.h(2)
circuit.x(1)

circuit.h([0, 1, 2])
circuit.x([0, 1, 2])

circuit.h(2)
circuit.ccx(0, 1, 2)
circuit.h(2)

circuit.x([0, 1, 2])
circuit.h([0, 1, 2])

circuit.measure([0, 1, 2], [0, 1, 2])

print(circuit.draw())

simulation = Aer.get_backend('aer_simulator')
compiled = transpile(circuit)
job = simulation.run(compiled, shots=1000)

results = job.results()
count = results.get_count()

print("results:", count)
```
